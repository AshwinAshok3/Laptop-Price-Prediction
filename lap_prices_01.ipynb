We Have to categorize these columns 

```python
price_col = laptop_df['Price_euros']
price_col
```

```python
laptop_df.drop(columns=['Price_euros'],inplace=True)
laptop_df.head(5)
```

<h3><center> Label Encodinng the Data </center></h3>

```python
le = LabelEncoder()
```

```python
def le_dfs(df, cols):
    for i, j in enumerate(cols):
        df[cols[i]] = le.fit_transform(df[cols[i]])
    return df
```

```python
new_laptop_df = le_dfs(laptop_df, list(laptop_df.columns))
new_laptop_df.head(5)
```

```python
new_laptop_df['Price_euros'] = price_col
new_laptop_df.head(5)
```

```python
# Now we check if there are any duplicate rows present in the dataset
print("Number Of Duplicate rows : ",new_laptop_df.duplicated().sum())

# If there is any missing values or null values present in the dataset
print("Number of Null values : ", new_laptop_df.isnull().sum().sum())
```

### V . Export the Dataset

```python
new_laptop_df.to_csv("encoded_laptop_prices1.csv")
```
